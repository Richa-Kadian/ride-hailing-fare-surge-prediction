```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv("rideshare_kaggle.csv")

# sampling
df_1 = df_1.sample(50000, random_state=42)

# 1 Missing values
for col in df_1.columns:
    df_1.loc[df_1.sample(frac=0.1).index, col] = np.nan

# 2 Noise
num_cols = df_1.select_dtypes(include=np.number).columns
for col in num_cols:
    df_1[col] = df_1[col] + np.random.normal(0, 5, df_1.shape[0])

# 3 Outliers
df_1.loc[df_1.sample(frac=0.05).index, num_cols] *= 10

# 4 Duplicate rows (LAST STEP)
duplicate_rows = df_1.sample(500)
df_1 = pd.concat([df_1, duplicate_rows], ignore_index=True)

# save
df_1.to_csv("dirty_dataset.csv", index=False)

print("Duplicate rows:", df_1.duplicated().sum())
```
